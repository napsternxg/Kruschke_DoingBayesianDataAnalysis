```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
sns.set_style("ticks")
sns.set_context("poster")
```

```python
x = np.array([1,2,3,4])
model_a = np.ones_like(x)/4.0
model_b = x/10.
model_c = 12./(25*x)


df = pd.DataFrame(np.vstack((model_a, model_b, model_c)), columns=x, index=["A", "B", "C"])
df
```

```python
width = 0.2
plt.bar(x, model_a, color="r", width=width, label="Model A")
plt.bar(x+width, model_b, color="b", width=width, label="Model B")
plt.bar(x+width*2, model_c, color="y", width=width, label="Model C")
plt.xlabel("Die Side")
plt.ylabel("$p(x)$")
plt.legend()
```

```python
case_1 = np.array([25,25,25,25])
case_1_p = case_1 * 1./ case_1.sum()

case_2 = np.array([48,24,16,12])
case_2_p = case_2 * 1./ case_2.sum()

width = 0.2
plt.bar(x, model_a, color="r", width=width, label="Model A")
plt.bar(x+width, model_b, color="b", width=width, label="Model B")
plt.bar(x+width*2, model_c, color="y", width=width, label="Model C")

plt.plot(x,case_1_p, "-ko", label="Case 1")
plt.plot(x,case_2_p, "-go", label="Case 2")
plt.xlabel("Die Side")
plt.ylabel("$p(x)$")
plt.legend(ncol=2, frameon=True, fancybox=True)
```

* Case 1 - Model 1 is more likely
* Case 2 - Model 3 is more likely
